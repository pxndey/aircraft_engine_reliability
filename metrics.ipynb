```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import gc
import pickle
```

```python
with open('history_array.pkl', 'rb') as f:
    history_arrays = pickle.load(f)

# 'history_arrays' will now be a list of dictionaries, each containing the training history for each model
window_size = [10,15,20,25,30,35,40,45,50]
```

```python
history_arrays[0]
```

```python
# Example: Plotting loss for the first model in the array
import matplotlib.pyplot as plt
def plot(history_arrays,window_size,i,):
    first_model_history = history_arrays[i]
    plt.plot(first_model_history['mse'])
    plt.plot(first_model_history['val_mse'])
    plt.title(f'Window Length {window_size[i]} Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])
    plt.show()
```

```python
plot(history_arrays,window_size,0)
```
